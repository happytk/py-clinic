1-100까지 숫자에 대해서 369게임을 만들어봅시다.

출력형식은 1,2,CLAP,4,5,CLAP,7,8,CLAP .... ,CLAP, CLAPCLAP, 100

```python
[
    ''.join(['clap' for ch in str(i) if ch in '369']) or i
    for i in range(1, 101)
]

```

```python
for i in range(1, 101):
    claps = []
    for ch in str(i):
        if ch in '369':
            claps.append('clap')
    
    if claps:
        print(''.join(claps))
    else:
        print(i)
```
