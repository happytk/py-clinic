1-100까지 숫자에 대해서 369게임을 만들어봅시다.

출력형식은 1,2,CLAP,4,5,CLAP,7,8,CLAP .... ,CLAP, CLAPCLAP, 100

```python
for i in range(1, 101):
    claps = []
    for number in str(i):
        if number in '369':
            claps.append('clap')
    print(''.join(claps) if claps else i)
```

```python
# 이렇게 줄여쓸 수도 있습니다. 중급과정에서 만나요!
for i in range(1, 101):
    print(''.join(['clap' for number in str(i) if number in '369']) or i)
```
