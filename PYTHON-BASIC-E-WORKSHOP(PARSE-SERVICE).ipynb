윈도우 시스템 내 서비스 포트를 관리하는 다음 파일 "c:/windows/system32/etc/services" 을 parsing하는 프로그램을 간단하게 작성해보겠습니다.

각 라인별로 처리해야 하며, 포맷은 다음과 같습니다.

```
<service name>  <port number>/<protocol>  [aliases...]   [#<comment>]
```

주석부분은 제거하고 처리합니다.

dict구조를 여러개 갖는 list 데이터 구조로 최종 결과를 정리합니다.

````python
lines = [
    line.strip() for line in open('services.txt')
    if line.strip() and not line.startswith('#')
]
````

````python
lines = [
    line.split('#')[0]
    for line in lines
]
````

````python
lines = [
    line.split(None, 2)
    for line in lines
]
````

````python
d = []
for line in lines:
    service_name = line[0]
    port, protocol = line[1].split('/')
    if len(line) > 2:
        alias = line[2]
    else:
        alias = ''
    d.append(dict(service_name=service_name,
                  port=port,
                  protocol=protocol,
                  alias=alias))
d
````
